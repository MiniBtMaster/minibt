```python
from minibt import *
options.set_conversion_mode="free"
```

```python
kline=LocalDatas.v2601_300.kline.iloc[:1000]
kline.head()
```

```python
from minibt.utils import get_pandas_explicit_methods
pandasta_method=list(get_pandas_explicit_methods(Factors))
for i in range(0, len(pandasta_method), 10):
    print(pandasta_method[i:i+10])
```

```python
rsi=kline.rsi()
zscore=kline.zscore()
macd=kline.macd()
adx=kline.adx()
pandasta_method.remove("adaptive_weight_trend")

for method in pandasta_method:
    ind=getattr(kline.close.factors,method)(rsi,zscore,macd,adx)
    print(f"指标{method},lines:{ind.lines},最后的数据:{ind.values[-1]}")

kline.close.factors.adaptive_weight_trend()
```
