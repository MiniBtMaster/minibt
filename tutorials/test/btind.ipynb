```python
kline=LocalDatas.v2601_300.kline.iloc[:1000]
kline.head()
```

```python
from minibt.utils import get_pandas_explicit_methods
pandasta_method=list(get_pandas_explicit_methods(BtInd))
for i in range(0, len(pandasta_method), 10):
    print(pandasta_method[i:i+10])
```

```python
default_method=[]
other=[]
for method in pandasta_method:
    try:
        ind=getattr(kline.btind,method)()
        print(f"指标{method},lines:{ind.lines},最后的数据:{ind.values[-1]}")
        default_method.append(method)
    except:
        other.append(method)
print("默认参数可运行指标:")
for i in range(0, len(default_method), 10):
    print(default_method[i:i+10])
```
