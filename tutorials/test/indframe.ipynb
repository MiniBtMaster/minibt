```python
from minibt import *
kline=LocalDatas.v2601_300.kline
kline.head()
```

```python
macd=kline.macd(height=300)
macd.tail(),type(macd),macd.source_object
```

```python
macd.btplot()
```

```python
sreies_x=macd.series.macdx
line_x=macd.macdx
type(sreies_x),type(line_x)
```

```python
(macd.long_signal
,macd.exitlong_signal
,macd.short_signal
,macd.exitshort_signal)
```

```python
macd.long_signal=macd.macds.cross_up(0)
macd.exitlong_signal=macd.macds.cross_down(0)
macd.short_signal=macd.macds.cross_down(0)
macd.exitshort_signal=macd.macds.cross_up(0)
macd.get_indicator_kwargs()
```

```python
macd.signal_style
macd.signal_style.long_signal=SignalStyle("low",Colors.cyan,Markers.circle_x)
macd.signal_key.short_signal="high"
macd.signal_show.short_signal=True
macd.signal_color.short_signal=Colors.yellow
macd.signal_overlap.short_signal=True
macd.signal_size.short_signal=20
macd.signal_label.short_signal=SignalLabel("sell")
macd.btplot()
```

```python
macd.line_style
macd.line_style=LineStyle()
macd.line_dash.macdx=LineDash.dashdot
macd.line_width.macdx=2.
macd.line_color.macdx=Colors.blue
macd.btplot()
```

```python
macd.to_lines()
macd.to_ndarray()
macd.select_dtypes(include=['number',])
macds=macd.pop("macds")
macd.get_indicator_kwargs(),macds.get_indicator_kwargs()
```
