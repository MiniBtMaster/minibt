```python
class MA(BtIndicator):
    params=dict(length1=3,length2=5)
    isplot = dict(long_signal=False, short_signal=False)
    overlap = dict(ma3=True, ma5=True)

    def next(self):
        ma3 = self.close.sma(self.params.length1)
        ma5 = self.close.sma(self.params.length2)
        long_signal = ma3.cross_up(ma5)
        short_signal = ma3.cross_down(ma5)
        return ma3, ma5, long_signal, short_signal
    
ma=MA(kline)
ma.btplot()
```
